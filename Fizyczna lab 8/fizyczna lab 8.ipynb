```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.constants import R
```

```python
class Substancja:
    def __init__(self, pusty, I, II): 
        self.kropla=((I-pusty)+(II-I))/20
```

```python
T=293.15 #[K]
woda=Substancja(15.5395, 16.1928, 16.8475)
oct1=Substancja(15.275, 15.8128, 16.3491)
oct2=Substancja(16.2139, 16.6945, 17.1761)
woda.sigma=72.74 #napięcie powierzchniowe cieczy wzorcowej w mN/m w 293.15 K
```

```python
oct1.sigma=woda.sigma*oct1.kropla/woda.kropla
oct2.sigma=woda.sigma*oct2.kropla/woda.kropla
sigma=[woda.sigma, oct1.sigma, oct2.sigma]
C=[0,1,2]
sigma
```

```python
Gamma1=-(1*(oct1.sigma-woda.sigma/1))/(R*T)
Gamma2=-(2*(oct2.sigma-oct1.sigma/1))/(R*T)
Gamma1, Gamma2
```
