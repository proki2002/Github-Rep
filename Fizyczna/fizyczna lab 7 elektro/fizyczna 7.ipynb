```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
logCu=np.array([-8,
-7,
-6,
-5,
-4,
-3,
-2,
-1])
ECu=np.array([151,
142,
133,
145,
187,
219,
230,
242])
ECu=ECu
pH=np.array([1,
2,
3,
4,
5,
6,
8,
9,
10])
EH=np.array([267,
227,
156,
96,
38,
-7.3,
-120.1,
-159,
-187])
a=np.polyfit(logCu[3:6],ECu[3:6],1)[0]
b=np.polyfit(logCu[3:6],ECu[3:6],1)[1]
r=np.polyfit(logCu[3:6],ECu[3:6],1,full=True)[1]
R2=1 - r / sum((ECu[3:6] - ECu[3:6].mean())**2) 
R2=float(R2)
```

```python
z=np.array([logCu[3:6][0]-0.55,logCu[3:6][-1]+1])
linia=np.mean(ECu[:4])
plt.scatter(logCu,ECu,color='k')
plt.plot(z,a*z+b,linestyle='dashed',label=f'R$^2$={round(R2,4)}')
plt.axhline(y=linia,linestyle='dashed')
plt.axvline(x=-5.12,color='r',linestyle=':', label='log$C_{Cu^{2+}}$=$-5.12$')
plt.xlabel('log$C_{Cu^{2+}}$ [-]')
plt.ylabel('E [mV]')
plt.legend()
plt.show()
```

```python
a=np.polyfit(pH,EH,1)[0]
b=np.polyfit(pH,EH,1)[1]
z=np.array([pH[0],pH[-1]])
r=np.polyfit(pH,EH,1,full=True)[1]
R2=1 - r / sum((EH - EH.mean())**2) 
R2=float(R2)
```

```python
plt.scatter(pH,EH,color='k')
plt.plot(z,z*a+b,linestyle='dashed',label=f'R$^2$={round(R2,4)}')
plt.xlabel('pH [-]')
plt.ylabel('E [mV]')
plt.legend()
plt.show()
```
