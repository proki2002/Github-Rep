```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
class UVVis:
    def __init__(self,xlsx):
        df=pd.read_excel(xlsx, header=8)
        df.dropna(axis='columns', inplace=True)
        lamb=[]
        A=[]
        for i in range(len(df.columns)):
            if (i+1)%2==1:
                lamb.append(i)
                lamb[int((i+1)/2)]=df[df.columns[i]].values.tolist()
            else:
                A.append(i)
                A[int(i/2)]=df[df.columns[i]].values.tolist()
        df=pd.read_excel(xlsx, header=3)
        df.dropna(axis='columns',inplace=True)
        label=df.columns.values.tolist()
        for x in label:
            if 'Sample' in x:
                label.pop(label.index(x))
        self.wyniki=[lamb,A,label]
        pass
def Wykres(uvvis1, uvvis2, rec):
    for i in range(len(uvvis1.wyniki[0])):
        plt.plot(uvvis1.wyniki[0][i],uvvis1.wyniki[1][i],label=uvvis1.wyniki[2][i])
        plt.plot(uvvis2.wyniki[0][i],uvvis2.wyniki[1][i],label=uvvis2.wyniki[2][i])
        plt.plot(rec[0],rec[1],label=rec[2])
        plt.xlabel('$\lambda$ [nm]')
        plt.ylabel('A [-]')
        plt.style.use('seaborn-whitegrid')
        plt.legend()
        plt.show()
```

```python
jak1=UVVis('irno2-mecn-10eq.xlsx')
jak1na=UVVis('irno2-mecn-10eq-1eqna.xlsx')
jak2=UVVis('irno2-mecn-10eq-2.xlsx')
jak2na=UVVis('irno2-mecn-10eq-1eqna-2.xlsx')
pomiary=[jak1,jak1na,jak2,jak2na]
for x in pomiary:
    for i in range(len(x.wyniki[0])):
        plt.plot(x.wyniki[0][i],x.wyniki[1][i],label=x.wyniki[2][i])
    plt.plot(rec[0],rec[1],label=rec[2])
    plt.xlabel('$\lambda$ [nm]')
    plt.ylabel('A [-]')
    plt.legend()
    plt.style.use('seaborn-whitegrid')
    plt.show()
rec=[jak1.wyniki[0][0], jak1.wyniki[1][0], jak1.wyniki[2][0]]
```

```python
Wykres(jak1,jak1na,rec)
Wykres(jak2,jak2na,rec)
```
