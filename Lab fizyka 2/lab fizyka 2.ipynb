```python
import numpy as np
from matplotlib import pyplot as plt 
from math import cos, tan
from scipy.constants import pi
```

```python
def LinReg(x,y):
  from sympy import sqrt
  X=np.array(x)
  Y=np.array(y)
  a=(np.count_nonzero(X)*np.sum(X*Y)-np.sum(X)*np.sum(Y))/(np.count_nonzero(X)*np.sum(X*X)-(np.sum(X)**2))
  b=(np.sum(Y)-a*np.sum(X))/(np.count_nonzero(X))
  S=sqrt((np.sum(Y*Y)-a*np.sum(X*Y)-b*np.sum(Y))/(np.count_nonzero(X)-2))
  da=sqrt(S**2*(np.count_nonzero(X))/(np.count_nonzero(X)*np.sum(X*X)-(np.sum(X))**2))
  db=sqrt(S**2*(np.sum(X*X))/(np.count_nonzero(X)*np.sum(X*X)-(np.sum(X))**2))
  return [a, b], [da, db]
```

```python
alpha=np.array([73.9,
75,
80,
85,
90,
95,
100,
105,
110,
115,
120,
125,
130,
135,
140,
145,
150,
155,
160])
alpha=alpha-80   #ustawiłem kąt przy którym I max jako 0 deg
I=np.array([355,
345,
360,
345,
330,
305,
275,
255,
230,
205,
175,
145,
120,
85,
55,
35,
20,
10,
5])
beta=alpha*pi/180
c=np.array(list(map(cos,beta)))
c=c**2
I_0=np.max(I)
I_0
```

```python
n=0
p=np.array([c[-1],c[2]])
plt.plot(p,I_0*p, color='k', label='Linia wyznaczona przez prawo Malusa')
k=LinReg(c,I)
plt.scatter(c,I, label='Punkty doświadczalne')
plt.plot([0,1],k[0][0]*np.array([0,1])+k[0][1], color='r', label='Linia regresji')
plt.xlabel('Kwadrat cosinusa kąta alpha [-]')
plt.ylabel('Natężenie prądu w fotokomórce $I$ [mA]')
plt.legend(loc='best')
plt.show()
k
```

```python
def Współczynnik_Załamania(a,b):
    kąt_B=abs(a-b)/2
    współ_zał=tan(kąt_B*pi/180)
    return współ_zał
```
